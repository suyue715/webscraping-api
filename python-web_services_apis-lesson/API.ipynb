```python
url = 'https://naics-codes.firebaseio.com/codes.json'
res = requests.get(url)
codes_dict = json.loads(res.content)
```

```python
codes_dict
```

```python
pd.DataFrame.from_dict(res.json(),orient='index')
```

```python
codes = []
for k, v in codes_dict.items():
    code = {}
    code['code'] = k
    code['description'] = v
    codes.append(code)
pd.DataFrame(codes)
```

```python
codes = [{'code': code, 'description': description} for code, description in codes_dict.items()]
pd.DataFrame(codes)
```
